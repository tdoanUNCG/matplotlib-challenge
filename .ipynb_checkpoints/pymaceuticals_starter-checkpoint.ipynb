```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = 'Resources/mouse_drug_data.csv'
clinical_trial_data_to_load = 'Resources/clinicaltrial_data.csv'

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinical_trial_data_to_load)
# Test print df
# mouse_data.head()
# clinical_data.head()

# Combine the data into a single dataset
complete_dataset = pd.merge(clinical_data,mouse_data,how='left',on=['Mouse ID','Mouse ID'])

# Display the data table for preview
complete_dataset.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
dataset_groupby = complete_dataset.groupby(['Drug','Timepoint'])
dataset_mean = dataset_groupby.mean()['Tumor Volume (mm3)']
# Convert to DataFrame
dataset_vol_mean = pd.DataFrame(dataset_mean)
dataset_vol_mean = dataset_vol_mean.reset_index()
# Preview DataFrame
dataset_vol_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
dataset_std_err = dataset_groupby.sem()['Tumor Volume (mm3)']
# Convert to DataFrame
dataset_std_err = pd.DataFrame(dataset_std_err)
dataset_std_err = dataset_std_err.reset_index()
# Preview DataFrame
dataset_std_err.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
dataset_vol_mean_pivot = dataset_vol_mean.pivot(index='Timepoint',columns='Drug')['Tumor Volume (mm3)']
dataset_std_err_pivot = dataset_std_err.pivot(index='Timepoint',columns='Drug')['Tumor Volume (mm3)']
# Preview that Reformatting worked
# dataset_std_err_pivot.head()
dataset_vol_mean_pivot.head()
```

```python
# Generate the Plot (with Error Bars)

plt.errorbar(dataset_vol_mean_pivot.index, dataset_vol_mean_pivot['Capomulin'],yerr=dataset_std_err_pivot['Capomulin'],color='r',marker='o',markersize=5,linestyle='dashed',linewidth=0.50)
plt.errorbar(dataset_vol_mean_pivot.index, dataset_vol_mean_pivot['Infubinol'],yerr=dataset_std_err_pivot['Infubinol'],color='b',marker='^',markersize=5,linestyle='dashed',linewidth=0.50)
plt.errorbar(dataset_vol_mean_pivot.index, dataset_vol_mean_pivot['Ketapril'],yerr=dataset_std_err_pivot['Ketapril'],color='g',marker='s',markersize=5,linestyle='dashed',linewidth=0.50)
plt.errorbar(dataset_vol_mean_pivot.index, dataset_vol_mean_pivot['Placebo'],yerr=dataset_std_err_pivot['Placebo'],color='k',marker='o',markersize=5,linestyle='dashed',linewidth=0.50)

plt.title('Tumor Response to Treatment')
plt.ylabel('Tumor Volume (mm3)')
plt.xlabel('Time (Days)')
plt.grid(True)
plt.legend(loc='best',fontsize='small',fancybox=True)

# Save the Figure
plt.savefig('Resources/figure_one')
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_site = dataset_groupby.mean()['Metastatic Sites']
# Convert to DataFrame
mean_met_site = pd.DataFrame(mean_met_site)
mean_met_site = mean_met_site.reset_index()
# Preview DataFrame
mean_met_site.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
std_err_met_site = dataset_groupby.sem()['Metastatic Sites']
# Convert to DataFrame
std_err_met_site  = pd.DataFrame(std_err_met_site)
std_err_met_site = std_err_met_site.reset_index()
# Preview DataFrame
std_err_met_site.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_met_site_pivot = mean_met_site.pivot(index='Timepoint',columns='Drug')['Metastatic Sites']
std_err_met_site_pivot = std_err_met_site.pivot(index='Timepoint',columns='Drug')['Metastatic Sites']
# Preview that Reformatting worked
#std_err_met_site_pivot.head()
mean_met_site_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(mean_met_site_pivot.index, mean_met_site_pivot['Capomulin'],yerr=std_err_met_site_pivot['Capomulin'],color='r',marker='o',markersize=5,linestyle='dashed',linewidth=0.50)
plt.errorbar(mean_met_site_pivot.index, mean_met_site_pivot['Infubinol'],yerr=std_err_met_site_pivot['Infubinol'],color='b',marker='^',markersize=5,linestyle='dashed',linewidth=0.50)
plt.errorbar(mean_met_site_pivot.index, mean_met_site_pivot['Ketapril'],yerr=std_err_met_site_pivot['Ketapril'],color='g',marker='s',markersize=5,linestyle='dashed',linewidth=0.50)
plt.errorbar(mean_met_site_pivot.index, mean_met_site_pivot['Placebo'],yerr=std_err_met_site_pivot['Placebo'],color='k',marker='d',markersize=5,linestyle='dashed',linewidth=0.50)

plt.title('Metastatic Spread During Treatment')
plt.ylabel('Meta Sites')
plt.xlabel('Treatment Duration (Days)')
plt.grid(True)
plt.legend(loc='best',fontsize='small',fancybox=True)

# Save the Figure
plt.savefig('Resources/figure_two.png')

# Show the Figure
plt.show()
```

![Metastatic Spread During Treatment](../Images/spread.png)

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_rate = dataset_groupby.count()['Tumor Volume (mm3)']
# Convert to DataFrame
survival_rate = pd.DataFrame({'Mouse Count': survival_rate})
survival_rate = survival_rate.reset_index()
# Preview DataFrame
survival_rate.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_rate_pivot = survival_rate.pivot(index='Timepoint',columns='Drug')['Mouse Count']

# Preview the Data Frame
survival_rate_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot((survival_rate_pivot['Capomulin']/25)*100,'ro',markersize=5,linestyle='dashed',linewidth=0.50)
plt.plot((survival_rate_pivot['Infubinol']/25)*100,'b^',markersize=5,linestyle='dashed',linewidth=0.50)
plt.plot((survival_rate_pivot['Ketapril']/25)*100,'gs',markersize=5,linestyle='dashed',linewidth=0.50)
plt.plot((survival_rate_pivot['Placebo']/25)*100,'kd',markersize=5,linestyle='dashed',linewidth=0.50)
plt.title('Survival During Treatment')
plt.ylabel('Survival Rate (%)')
plt.xlabel('Time (Days)')
plt.grid(True)
plt.legend(loc='best',fontsize='small',fancybox=True)

# Save the Figure
plt.savefig('Resources/figure_three.png')

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_change = 100*(dataset_vol_mean_pivot.iloc[-1] - dataset_vol_mean_pivot.iloc[0])/dataset_vol_mean_pivot.iloc[0]
percent_change_std_err = 100*(dataset_std_err_pivot.iloc[-1] - dataset_std_err_pivot.iloc[0])/dataset_std_err_pivot.iloc[0]
# percent_change = pd.DataFrame(percent_change)
# percent_change_std_err = pd.DataFrame(percent_change_std_err)
# Display the data to confirm
percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
rel_perc_change = (percent_change['Capomulin'],
                  percent_change['Infubinol'],
                  percent_change['Ketapril'],
                  percent_change['Placebo'])

# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
ind = np.arange(len(rel_perc_change))  
width = 1
pass_rect = ax.bar(ind[0], rel_perc_change[0], width, color='green')
fail_rect = ax.bar(ind[1:], rel_perc_change[1:], width, color='red')

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_xticks(ind + 0.5)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
ax.set_autoscaley_on(False)
ax.set_ylim([-30,70])
ax.grid(True)

# Use functions to label the percentages of changes
def autolabelFail(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 3,
                '%d%%' % int(height),
                ha='center', va='bottom', color="white")

def autolabelPass(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., -8,
                '-%d%% ' % int(height),
                ha='center', va='bottom', color="white")

# Call functions to implement the function calls
autolabelPass(pass_rect)
autolabelFail(fail_rect)

# Save the Figure
fig.savefig("Resources/figure_four.png")

# Show the Figure
fig.show()
```
